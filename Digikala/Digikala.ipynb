```python
import pandas as pd
import numpy as np
import itertools
import timeit
```

```python
df=pd.read_csv('train.csv')
```

```python
from ast import literal_eval
d=[]
c=[]
for i in range(0,68839):
    s=literal_eval(df['product_description'][i])
    d.append(list(s.keys()))

f=list(itertools.chain.from_iterable(d))
r=set(f)
```

```python
w=list(r)
w.append('price')
```

```python
data=pd.DataFrame(columns=w)
data
```
