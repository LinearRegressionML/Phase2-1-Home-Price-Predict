```python
import sqlite3
import pandas as pd
```

```python
df=pd.read_csv('merged2.csv')
df= df.loc[:,~df.columns.str.contains('^Unnamed: 0')]
df.head()
```

```python
conn = sqlite3.connect('Home_database_IranFile') 
c = conn.cursor()

c.execute('''
          CREATE TABLE IF NOT EXISTS home
          ([home_id] INTEGER PRIMARY KEY, [construct_year] INTEGER, [number_of_room] INTEGER,
            [Elevator] Boolean, [Parking] boolean, [size_metre] integer, [price] integer,
            [main_neighbourhood] string, [Cage] boolean, [balcon] boolean[floor] integer)
          ''')
                     
conn.commit()
```

```python
df.to_sql('home',conn,if_exists='replace',index=False)
conn.commit()
conn.close()
```
