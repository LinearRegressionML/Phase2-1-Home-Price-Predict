```python
import sqlite3
import pandas as pd
```

```python
df=pd.read_csv('FINAL_CLEAN2.csv')
df= df.loc[:,~df.columns.str.contains('^Unnamed: 0')]
df.head()
```

```python
conn = sqlite3.connect('Home_database') 
c = conn.cursor()

c.execute('''
          CREATE TABLE IF NOT EXISTS home
          ([home_id] INTEGER PRIMARY KEY, [year] INTEGER, [tabaghat] INTEGER,
            [parking] Boolean, [anbari] boolean, [metrag] integer, [price] integer,
            [source] string, [asansor] boolean, [room] integer, [floor] integer, [lat] float, [long] float,
            [vahed] integer, [shomali] integer, [date] Date)
          ''')
                     
conn.commit()
```

```python
df.to_sql('home',conn,if_exists='replace',index=False)
conn.commit()
conn.close()
```
