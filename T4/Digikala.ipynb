```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
# for i in df:
#     print(i.product_description)
df
```

```python
from ast import literal_eval
```

```python
for i in range(0,10):
    s1=literal_eval(df['product_description'][i])
    dw=pd.DataFrame( [[df.id[i],s1,df.price[i]]],
                    columns=['id', 'product_description','price'])

    d=pd.concat([dw.drop(['product_description'], axis=1), dw['product_description'].apply(pd.Series)], axis=1)
    d.to_csv('ok.csv')
    print(d)
```
